## Data Cleaning ##

```python
def clean_price(price):
    if isinstance(price,str):
        if '£' in price:
            price = price.replace('£','').replace('each est.','').replace('-','').strip()
        elif 'p' in price:
            price = '0.' + price.replace('p','').replace('each est.','').replace('-','').strip()

    return float(price)
```

```python
mid_df['price'] = mid_df['item-price'].apply(clean_price)
```

```python
# Example DataFrame
data = {'name': ['Item1', 'Item2', 'Item3'],
        'price': [100, 200, 150],
        'size': ['4x18 g', '250ml', '500g']}
cream = pd.DataFrame(data)

# Function to preprocess the `size` column
def preprocess_size(size):
    size = size.lower().strip()
    if 'x' in size:  # Handle cases like '4x18 g'
        parts = size.split('x')
        size = float(parts[0]) * float(parts[1].replace('g', '').replace('ml', '').strip())
    else:
        size = float(size.replace('g', '').replace('ml', '').strip())
    return size

# Apply preprocessing to create a numeric `size` column
cream['size_numeric'] = cream['size'].apply(preprocess_size)

# Calculate `new_price` per 1000 units
cream['new_price'] = cream['price'] / cream['size_numeric'] * 1000

# Display selected columns
print(cream[['name', 'price', 'size', 'new_price']])
```

```python
cream= (
    cream.assign(new_price=lambda s: s['price'] / s['size'].str.replace('g', ' ', regex=True).str.replace('ml', ' ', regex=True).astype('float') * 1000)
)

cream[['name', 'price', 'size', 'new_price']]
```

```python
group_cream = cream.groupby('name')['new_price'].mean().sort_values(ascending=False).to_frame().reset_index()
group_cream
```

```python
(
    ggplot(group_cream, aes(x='name', y='new_price')) +
    geom_area(color='red', size=1.2) +
    labs(
        title='Price of Corn Flakes',
        x='Product Name',
        y='Price ($)'
    ) +  
    theme(
        axis_text_x=element_text(hjust=1),  
        plot_title=element_text(size=14, face='bold')  
    ) + ggsize(600, 800)
)
```
